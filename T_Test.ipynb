```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("Preproc_Placement.csv")
```

```python
dataset
```

```python
#Independent Sample - Unpaired T-Test
#Different Group(Male,Female) and same condition(salary)
```

```python
from scipy.stats import ttest_ind
male_sal=dataset[dataset['gender']=='M']['salary']
female_sal=dataset[dataset['gender']=='F']['salary']
ttest_ind(male_sal,female_sal)
```

```python
#Independent Sample - Unpaired T-test
#Different Group(hsc_board - others & central) and same condition(hsc_p)
```

```python
from scipy.stats import ttest_ind
b_others=dataset[dataset["hsc_b"]=='Others']["hsc_p"]
b_central=dataset[dataset["hsc_b"]=="Central"]["hsc_p"]
ttest_ind(b_others,b_central)
```

```python
#Independent Sample - Unpaired T-test
#Different Group(hsc_s - Commerce & Science) and same condition(mba_p)
from scipy.stats import ttest_ind
com_hsc=dataset[dataset["hsc_s"]=='Commerce']["mba_p"]
sci_hsc=dataset[dataset["hsc_s"]=="Science"]["mba_p"]
ttest_ind(com_hsc,sci_hsc)
```

```python
#Dependent Sample - Paired T-Test
#Same Group(Male) and Different condition(ssc_p,hsc_p)
```

```python
from scipy.stats import ttest_rel
sscp=dataset[dataset["gender"]=='M']['ssc_p']
hscp=dataset[dataset["gender"]=='M']['hsc_p']
ttest_rel(sscp,hscp)
```

```python
#Dependent sample - Paired T-Test
#Same Group(Commerce) and different condition(etest_p,mba_p)
```

```python
from scipy.stats import ttest_rel
com_etest=dataset[dataset["hsc_s"]=="Commerce"]["etest_p"]
com_mba=dataset[dataset["hsc_s"]=="Commerce"]["mba_p"]
ttest_rel(com_etest,com_mba)
```

```python
#Dependent sample - Paired T-Test
#Same Group(Science) and different condition(etest_p,mba_p)
from scipy.stats import ttest_rel
sci_etest=dataset[dataset["hsc_s"]=="Science"]["etest_p"]
com_mba=dataset[dataset["hsc_s"]=="Science"]["mba_p"]
ttest_rel(sci_hsc,com_hsc)
```
