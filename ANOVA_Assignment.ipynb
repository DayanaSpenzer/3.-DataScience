```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
#ANOVA ONE WAY CLASSIFICATION
import scipy.stats as stats
stats.f_oneway(dataset['gender'],
               dataset['degree_t'],
               dataset['salary'])
```

```python
#TWO WAY ANOVA CLASSIFICATION
import numpy as np 
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
#H0: 
#1. The gender doesnt make any impact on the salary 
#2. The specialisation doesnt make any impact on the salary
#3. The interaction of gender and specialisation doesnt make any impact on the salary
```

```python
Gen=dataset['gender']
Spec=dataset['specialisation']
Sal=dataset['salary']
model = ols('Sal ~ Gen + Spec + Gen:Spec', data=dataset).fit()
anova_result = sm.stats.anova_lm(model, type=2)
```

```python
anova_result
```

```python
#The null hypothesis is accepted as all the p value is greater than 0.05
#ie, The gender, specialisation and the interaction of gender and 
#specialsiation doesnt make any impact on salary
```
